# **Pandas Series**
Pandas series are 1d labeled array think of it like a single column

```python
data=[100,200,300]
series=pd.Series(data)
print(series)
```

```python
series=pd.Series(data,index=['a','b','c'])
print(series)
```

```python
print(series.loc['a'])
```

```python
print(series.loc['b'])
```

```python
print(series.loc['c'])
```

```python
series.loc['d']=500
print(series)
```

```python
print(series.iloc[0])
```

```python
print(series.iloc[1])
```

```python
data1=[100,200,300,400,566,589,256,349,645,120]
series=pd.Series(data1,index=['a','b','c','d','e','f','g','h','i','j'])
print(series[series>200])
```

```python
calories={
    "Day 1":1750,
    "Day 2":500,
    "Day 3":1700
}
series=pd.Series(calories)
print(series)
```

```python
print(series.loc['Day 1'])
```

```python
series.loc['Day 2']+=500
print(series)
```

# **Data Frames**
A tablular data structure with rows and columns

```python
data={
    "Name":["abc","xyz","efg"],
    "Age":[21,31,41]
}
df=pd.DataFrame(data,index=['a','b','c'])
print(df)
```

```python
print(df.loc['a'])
```

```python
print(df.iloc[2])
```

```python
#add new column
df["Job"]=["dniri","jdfuh","fuegfu"]
print(df)
```

```python
#add new row
new_row=pd.DataFrame([{
    "Name":"sandy",
    "Age":41,
    "Job":"ehirhi"
}])
df=pd.concat([df,new_row])
print(df)
```

# **Aggregate Functions**

```python
print(df["Survived"].mean())
print(df["Survived"].sum())
print(df["Survived"].min())
print(df["Survived"].max())
print(df["Survived"].count())
```

```python
print(df.mean(numeric_only=True))
print(df.sum(numeric_only=True))
print(df.min(numeric_only=True))
print(df.max(numeric_only=True))
print(df.count(numeric_only=True))
```

# **Data Cleaning Using Pandas**

```python
#dropping col
df1=df.drop(columns=["SibSp" , "Parch"])
print(df1[0:6].to_string())
```

```python
#handle missing data
df2=df.dropna(subset=["Cabin"])
print(df2.to_string())
```

```python
#fill missing data
df3=df.fillna({"Cabin":"None"})
print(df3.to_string())
```

```python
#fix inconsistent values
df4=df["Sex"].replace({"female":"F","male":"M"})
print(df4.to_string())
```

```python
#standadize text
df5=df["Name"].str.lower()
print(df5.to_string())
```

```python
#fix datatypes
df["Survived"]=df["Survived"].astype(bool)
print(df.to_string())
```

```python
#remove duplicate values
df=df.drop_duplicates()
print(df.to_string())
```
