# **Python-operators**

```python
#Arithmetic Operators
#+,-,*,/,%,**,//
x=50
y=20
z=x+y
print(z)
z=x-y
print(z)
z=x*y
print(z)
z=x/y
print(z)
z=x%y
print(z)
z=x**y #exponentiation
print(z)
z=x//y #floor division
print(z)
```

```python
#Assignment Operators
# =,+=,-=,*=,/=,%=,//=,**=,&=,|=,^=,>>=,<<=,:=
x=10
y=5
x+=y
print(x)
x=10
y=5
x-=y
print(x)
x=10
y=5
x*=y
print(x)
x=10
y=5
x/=y
print(x)
x=10
y=5
x%=y
print(x)
x=10
y=5
x//=y
print(x)
x=10
y=5
x**=y
print(x)
x=10
y=5
x**=y
print(x)
x=10
y=5
x&=y
print(x)
x=10
y=5
x|=y
print(x)
x=10
y=5
x^=y
print(x)
x=10
y=5
x>>=y
print(x)
x=10
y=5
x<<=y
print(x)
x=10

print(x := 5)#Walrus Operator
#Walrus Operator-assigns a value and returns the value at he same time


```

```python
#Ternary Operator
num=6
x="WEEKEND" if num>5 else "WORKDAY"
print(x)
```

```python
#Comparision Operators
# ==,!=,>,<,>=
x=3
y=5
print(x==y)
print(x!=y)
print(x>y)
print(x<y)
print(x>=y)
print(x<=y)
```

```python
#Logical operators
#and,or,not
x=10
if(x>0 and x<20):
  print("with in range")
else:
  print('error')
if(x==0 or x==20):
  print("yes")
else:
  print("no")
print(not(x>0 and x<20))
```

```python
#identity operators
#is , is not
x=["gayathri","boina"]
y=["gayathri" , "boina"]
z=x
print(x is y)
print(y is x)
print(x is z)
print(z is x)
```

```python
#Membership Operators
# in, not in
fruits=(10,20,30,40)
print(80 in fruits)
print(80 not in fruits)
```

```python
#Bitwise Operators
# &,|,^,>>,<<,~
x=10
y=5
x&=y
print(x)
x=10
y=5
x|=y
print(x)
x=10
y=5
x^=y
print(x)
x=10
y=5
x>>=y
print(x)
x=10
y=5
x<<=y
print(x)
x=10
print(~x)# ~x=-(x+1)
```

# **Python Strings**

```python
fname="gayathri "
lname='boina'
print(fname+lname)

#quotes inside quotes -just know that outer quotes and inner quotes should not be matching
print("I'm Gayathri")

#in python strings are arrays there is no such datatype as char a single character s considered to be string of size 0
print(fname[5])

#looping
for x in fname:
  print(x)

#string length
print(len(fname))

print("a" in fname)
print("a" not in fname)
```

```python
#keyword arguments- with these order doesnt matter

def func(number,character):
  print(number)
  print(character)
func(character='c',number=10)
```

```python
#positional arguments
def func(val1,val2):
  print("printing val1 and it should be 10",val1)
  print("printing val2 and it should be 20",val2)
func(10,20)#position matters
```

```python
#if u want to write a function that only takes positional arguments
def func(val1,val2,/):
  print("printing val1 and it should be 10",val1)
  print("printing val2 and it should be 20",val2)
func(10,20)
```

```python
#if u want to write a function that only takes key-word arguments
def func(*,val1,val2):
  print("printing val1 and it should be 10",val1)
  print("printing val2 and it should be 20",val2)
func(val2=20,val1=10)
```

```python
#Python *args and **kwargs
#these allow functions to accept unkown number of arguments
def func(*args):#args allows us to pass any number of positional arguments its not only args u can keep * in the start and wite any name in place of args
  print(args[0])
  print(args[1])
  print(args[2])
tuple=(10,20,30)
func(tuple,tuple,tuple)

```

```python
# combining regular arguments and args remeber always regular once should come befor args
def func(a,b,*args):
  print(a)
  print(b)
  print(args)#tuple
  print(type(args))
func(10,20,30,40,60,5,8,9,77,55,69)
```

```python
#kwargs allow us to pass any number of key-word arguments just write ** before kwargs
def func(**kwargs):
  print(kwargs)
  print(type(kwargs))
func(a=10,b=20,c=30)
#can combine kwargs with regular arguments but regular should come first
```
