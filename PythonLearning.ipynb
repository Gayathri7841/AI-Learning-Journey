# **Rules for variable naming:-**
1) variable name must start with a letter or an underscore
2) variable name must only contain alpha-numeric characters and underscore
3) variable names are case-sensitive
4) variable cannot be a python keyword



```python
#multiple values to multiple variables
x,y,z=10,"gayathri",20
print(x,y,z)
```

```python
#one value to multiple variables
x=y=z=10
print(x,y,z)
```

```python
#unpacking a collection
fruits=["oranges","apples"]
a=b=fruits
print(a,b)
```

```python
fruits=["oranges","apples"]
a,b=fruits
print(a,b)
```

```python
#Global variables creation
x=10#global variable
def fn():
  y=20#local variable
  print(x,y)
print(x)
fn()
```

```python
#changing a local variable to global by using global keyword
def fun():
  global k
  k=45
  print(k)
print(k)
fun()
```
