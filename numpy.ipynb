# **Filtering**

```python
ages=np.array([[21,17,41,16,22,44,77,55],[39,25,55,53,46,31,18,19]])
teenagers=ages[ages<18]
adults=ages[ages>=18]
adults1=ages[(ages>=18)&(ages<=65)]
print(adults1)
print(adults)
print(teenagers)
```

# **Random Numbers**

```python
rng=np.random.default_rng()
```

```python
print(rng.integers(1,7))
```

```python
print(rng.integers(low=1,high=107,size=4))
```

```python
print(rng.integers(low=1,high=107,size=(3,2)))
```

```python
print(np.random.uniform())
np.random.seed(seed=1)# u will get same values if u use the same seed
print(np.random.uniform(low=-1,high=2,size=5))
```
