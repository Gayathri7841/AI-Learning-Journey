# Creating a basic matplotlib plot
1)we use plot() function to plot the coordinates on x and y axis

2)we use the function show() to display the plot

```python
x=[2023,2024,2025,2026]
y=[15,25,31,44]
plt.plot(x,y)
plt.show()
```

# Plot Customization

```python
x=np.array([25,65,85,77,49])
y1=np.array([1,4,3,7,8])
y2=np.array([7,8,12,4,1])

line_style=dict(
    marker="o",
    markersize=10,
    markerfacecolor="black",
    markeredgecolor="red",
    linestyle="solid",
    linewidth=3,
    color="green"
)
plt.plot(x,y1,**line_style);
plt.plot(x,y2,marker="o",markersize=10,markerfacecolor="black",markeredgecolor="red",linestyle="solid",linewidth=3,color="blue");
plt.show()
```

# Grids

```python
x=[1,2,3,4,5]
y=[5,10,15,20,25]
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
from matplotlib.lines import lineStyles
x=[1,2,3,4,5]
y=[5,10,15,20,25]
plt.grid(axis="y",linewidth=3,color="lightgray",linestyle="dashed")
plt.plot(x,y)
plt.show()
```

# Bar Chart

```python
categories=np.array(["Grains","Fruits","Vegetables","Proteins"])
values=np.array([4,3,2,5])
plt.bar(categories,values,color="skyblue")
plt.title("Food Consumption")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

# Pie Charts

```python
categories=np.array(["Grains","Fruits","Vegetables","Proteins"])
values=np.array([4,3,2,5])
colors=["red","skyblue","yellow","lightgreen"]
plt.title("Random Pie Chart")
plt.pie(values,labels=categories,autopct="%1.1f%%",colors=colors,explode=[0,0,0,0.1],shadow="true")
plt.show()
```

# Scatter Plots

```python
x=[0,1,1,2,3,4,5,6,7,7,8]
y1=[55,60,65,62,68,70,75,78,82,85,87]
y2=[77,55,44,66,88,99,12,33,44,22,11]
plt.scatter(x,y1,color="red",alpha=0.5,s=100,label="Class A")
plt.scatter(x,y2,color="blue",alpha=0.5,s=100,label="Class B")
plt.legend()
plt.show()
```

# Histograms

```python
scores=np.random.normal(loc=80,scale=10,size=100)
scores=np.clip(scores,0,100)
plt.title("Student scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.hist(scores,bins=10,color="red",edgecolor="blue")

plt.show()
```

# Subplots

```python
x=np.array([1,2,3,4,5])

fig,axes=plt.subplots(2,2)
axes[0,0].plot(x,x*2)
axes[0,0].set_title("x*2")
plt.show()
```
