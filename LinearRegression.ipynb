# **5)Feature Engineering**

```python
x['Date']=pd.to_datetime(x['Date'],errors='coerce',dayfirst=True)
x['SoldYear']=  x['Date'].dt.year
x['SoldMonth']=  x['Date'].dt.month
x['SoldDate']=  x['Date'].dt.day
x.drop(columns=['Date'],axis=1,inplace=True)
x.head(3)
```

# **6)Seperate Numerical and Categorical data**

```python
num=x.select_dtypes(include="number")
cat=x.select_dtypes(exclude="number")
```

```python
num
```

```python
cat
```

# **7) Preprocessing Pipeline(impute +encode + scaling)**

```python
#numeric value-median impute + scale
num_transformation=Pipeline(steps=[("imputer",SimpleImputer(strategy="median")),
                ("scaling",StandardScaler())])
#categorical :most frequent+one hot encodeing
cat_transformtion=Pipeline(steps=[("imputer",SimpleImputer(strategy="most_frequent")),
                ("onehot",OneHotEncoder(handle_unknown='ignore'))])
preprocessor=ColumnTransformer(transformers=[("num",num_transformation,num),("cat",cat_transformtion,cat)],
                  remainder="drop")
preprocessor
```
